```python
import utm


def to_utm(lon_lat):

  if type(lon_lat)== type(()) or type(lon_lat)==type([]) or type(lon_lat)==type(np.array([])):
    lat, lon = lon_lat
  else:
    return "Please give a tuple object of latitude and longitude"


  easting, northing = utm.from_latlon(lat, lon)[:2]

  return easting, northing


def cal_u_v(velocity_easting1_easting2_northing1_northing2):

  arg = velocity_easting1_easting2_northing1_northing2

  if type(arg)== type(()) or type(arg)==type([]) or type(arg)==type(np.array([])):

    velocity, easting1, easting2, northing1, northing2 = arg
  else:
    return "Please insert a tuple of river_discharge, easting1, easting2, northing1 and northing2"

  u = velocity*np.cos(np.arctan((easting1-easting2)/(northing1-northing2)))


  v = velocity*np.sin(np.arctan((easting1-easting2)/(northing1-northing2)))


  return u,v
```
